```python
import os , cv2 , numpy as np

fnotas = 'ucf_annotations.txt'
notas = []
with open(fnotas, 'r') as file:
    for line in file:
        notas.append(line.strip("\n"))
print(len(notas))
#for nota in notas : print(nota)
```

```python
vlist = list(open("test.list"))
print(len(vlist))
```

```python
def get_video_length(video_path):
    _ext = ['.avi', '.mp4']
    _, ext = os.path.splitext(video_path)
    if not ext in _ext:raise ValueError('Extension "%s" not supported' % ext)
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():raise ValueError("Could not open the file.\n{}".format(video_path))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    tframes = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.release()
    return tframes , fps
```

```python
GT = []
TFRAMES = 0

## iterates over all videos in test
for j , v in enumerate(vlist):
    v=v.strip("\n")
    fn=os.path.basename(v)
    print("\n",j,fn)
    
    for k , nota in enumerate(notas):
        nota=nota.split("  ")
        notafn=nota[0]
        notafintervls = nota[2:6]
        
        if notafn in fn:
            print(notafn,notafintervls)
            
            tframes,*_ = get_video_length(v)
            extra = tframes % 16
            if extra != 0 : tframes -= extra
            TFRAMES += tframes
            
            ## checks if frist video features len match
            #fnnpy = "/raid/DATASETS/anomaly/UCF_Crimes/features/C3DSPORTS1M/0000/test/Abuse028_x264.npy"
            #feat = np.load(fnnpy)
            #tsteps = feat.shape[0]
            #print(tsteps*16 , tframes)
            #print(fn , lfn ,"\n", tframes , fintervals)
            
            gt = np.zeros(tframes)

            ## only care for intervals pairs != -1
            if not all( finter == -1 for finter in notafintervls): 
                
                for i in range(0, len(notafintervls), 2):
                    sf, ef = int(notafintervls[i]), int(notafintervls[i + 1])
                    if sf == -1 or ef == -1: continue
                    gt[sf-1:ef] = 1
                
            GT.append((fn,gt))
            #GT.extend(gt) ## ALL
            print(tframes,len(gt))
            
            ## pica a b.owl
            if tframes < 200:
                print(fn , notafn ,"\n", tframes , notafintervls)
                for j,ff in enumerate(gt):
                    print(j+1,"  ",ff)
            
            break


print(TFRAMES , len(GT) , GT[0])
tframes = 0
for gt in GT:tframes += len(gt[1])
assert TFRAMES  == tframes
GT = np.array(GT,dtype=object)
np.save("gt-ucf_16f.npy",GT)

## ALL
#print(TFRAMES , len(GT))
#assert TFRAMES  == len(GT)
#np.save("gt-ucf_16f_all.npy",GT)
```
